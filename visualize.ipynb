```python
import numpy as np
import matplotlib.pyplot as plt
import os
from pathlib import Path
```

```python
SUBSET_ROOT = "subset_fixed"
SPLIT = "train"     # or "test"
```

```python
base_path = Path(SUBSET_ROOT) / SPLIT

feature_dirs = {
    "power_i": base_path / "power_i",
    "power_s": base_path / "power_s",
    "power_sca": base_path / "power_sca",
    "power_all": base_path / "power_all",
    "power_t": base_path / "power_t",
}

label_dir = base_path / "IR_drop"
```

```python
sample_name = next(iter(feature_dirs["power_i"].iterdir())).name
sample_name
```

```python
power_i   = np.load(feature_dirs["power_i"]   / sample_name)
power_s   = np.load(feature_dirs["power_s"]   / sample_name)
power_sca = np.load(feature_dirs["power_sca"] / sample_name)
power_all = np.load(feature_dirs["power_all"] / sample_name)
power_t   = np.load(feature_dirs["power_t"]   / sample_name)

label = np.load(label_dir / sample_name)
```

```python
fig, axs = plt.subplots(1, 5, figsize=(18,18))

axs[0].imshow(power_i.squeeze(), cmap="hot")
axs[0].set_title("Power I")

axs[1].imshow(power_s.squeeze(), cmap="hot")
axs[1].set_title("Power S")

axs[2].imshow(power_sca.squeeze(), cmap="hot")
axs[2].set_title("Power SCA")

axs[3].imshow(power_all.squeeze(), cmap="hot")
axs[3].set_title("Power ALL")

axs[4].imshow(label.squeeze(), cmap="hot")
axs[4].set_title("IR Drop")

plt.show()
```
